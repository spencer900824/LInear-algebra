```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# 先定義出S

```python
x = np.linspace( -10 , 10 , 21)
y = np.linspace( -10 ,10 , 21)
xx , yy = np.meshgrid(x,y)
xx,yy = xx.flatten() , yy.flatten()
S = np.vstack([xx,yy])

plt.scatter(xx,yy)
```

# 定義線性轉換Phi
    Phi : R**2 -> R**2
    Phi(x,y) = (5x - 7y + 6x - 3y)
    相對於R**2的標準基底
    Phi的矩陣表示法: A = [[5,6],[-7,-3]]
    

```python
A = np.array([[5,6]
            , [-7,-3]])
def phi(x):
    assert A.shape[1]==x.shape[0], 'Incompatible dimension'
    return A@x
```

# 經過線性轉換後的S

```python
i,j = phi(S)
plt.scatter(i,j)
```

# 比較看看S和Phi(S)

```python
plt.scatter(i,j,c='b')
plt.scatter(xx,yy,alpha=0.5,c='r')
```

# S長大了呢!

# 再試試另一種轉移矩陣

```python
B = np.array( [[ np.cos(np.pi/4),-1*np.sin(np.pi/4)],
                [np.sin(np.pi/4),np.cos(np.pi/4)]])
def tur(x):
    assert B.shape[1]==x.shape[0], 'Incompatible dimension'
    return B@x
```

# 轉動了!

```python
u,v = tur(S)
plt.scatter(u,v)
```

# 比較看看

```python
plt.scatter(u,v,c='b')
plt.scatter(xx,yy,alpha=0.5,c='r')
```
