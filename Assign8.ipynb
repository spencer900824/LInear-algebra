```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from sympy import pprint
sp.init_printing()
```

$x$ is a vector on $R^2$ , and $U$ is a subspace of $R^2$

$L = x+U$ is a Affine subspace of $R^2$

```python
x = np.array([2,2])
U = np.array([7,8])
a = np.linspace(-10,10,1000)
Lx = []
Ly = []

for i in a:
    p = x+np.dot(i,U)
    plt.scatter(p[0] , p[1] ,s=3, c='blue')
    Lx.append(p[0])
    Ly.append(p[1])
plt.axvline(0, c='black')
plt.axhline(0, c='black')
L = np.array([Lx,Ly]).T
```

$\Phi$ is a linear mapping of $R^2 $to$ R^2$ and $a$ is a  translation on $R^2$

```python
Phi = np.array([[-2,-4/3],
                [5,-3/7]])
a = np.array([18,27])
Vx=[]
Vy=[]
for i in range(L.shape[0]):
    w = a + np.dot(Phi,L[i])
    plt.scatter(w[0],w[1],s=3,c='red')
    Vx.append(w[0])
    Vy.append(w[1])
plt.axvline(0, c='black')
plt.axhline(0, c='black')  
V = np.array([Vx,Vy]).T
```

See if $L$ and $U$ are on same hyper plane

```python
plt.axvline(0, c='black')
plt.axhline(0, c='black')
plt.scatter(Lx, Ly, s=1, c = 'blue')
plt.scatter(Vx, Vy, s=1, c = 'red')
```
